### Data loading

### Subtask:
Load the data from "personal_data.csv" and "loan_data.csv" into pandas DataFrames.


**Reasoning**:
Import pandas and load the two CSV files into dataframes as requested.



```python
import pandas as pd

personal_df = pd.read_csv("personal_data.csv")
loan_df = pd.read_csv("loan_data.csv")

print("First few rows of personal_df:")
display(personal_df.head())

print("\nFirst few rows of loan_df:")
display(loan_df.head())
```

## Data exploration

### Subtask:
Explore the loaded dataframes. This includes checking the shape of the data, identifying columns, data types, and looking for missing values. Perform basic descriptive statistics on relevant columns.


**Reasoning**:
Perform initial exploration of both dataframes by checking their shape, data types, missing values, and basic statistics for numerical and categorical columns as requested by the instructions.



```python
print("Shape of personal_df:", personal_df.shape)
print("Shape of loan_df:", loan_df.shape)

print("-" * 30)

print("Info of personal_df:")
personal_df.info()

print("-" * 30)

print("Info of loan_df:")
loan_df.info()

print("-" * 30)

print("Missing values in personal_df:")
print(personal_df.isnull().sum())

print("-" * 30)

print("Missing values in loan_df:")
print(loan_df.isnull().sum())

print("-" * 30)

print("Descriptive statistics for personal_df (numerical columns):")
display(personal_df.describe())

print("-" * 30)

print("Descriptive statistics for loan_df (numerical columns):")
display(loan_df.describe())

print("-" * 30)

print("Value counts for 'gender' in personal_df:")
display(personal_df['gender'].value_counts())

print("-" * 30)

print("Value counts for 'status' in loan_df:")
display(loan_df['status'].value_counts())
```

## Data wrangling

### Subtask:
Merge the `personal_df` and `loan_df` DataFrames on the 'person_id' column to create a unified dataset for analysis. Before merging, ensure the 'person_id' columns in both dataframes are of the same data type if they aren't already.


**Reasoning**:
Check the data types of the 'person_id' column in both dataframes and merge them after ensuring consistent data types.



```python
print("Data types in personal_df:")
personal_df.info()

print("\nData types in loan_df:")
loan_df.info()


merged_df = pd.merge(personal_df, loan_df, on='person_id', how='inner')

print("\nFirst few rows of the merged DataFrame:")
display(merged_df.head())

print("\nShape of the merged DataFrame:", merged_df.shape)
```

## Data analysis

### Subtask:
Conduct a 4-point analysis on the merged dataset (`merged_df`). This analysis should explore relationships between personal attributes and loan characteristics to uncover hidden patterns. Based on the data exploration and merging steps, focus on the following:

1.  Analyze the distribution of `amount`, `tenure`, and `dpd` across different `gender` categories.
2.  Examine the distribution of `amount`, `tenure`, and `dpd` based on `status` of the loan (active, closed, default).
3.  Investigate the relationship between `tenure` and `amount`.
4.  Analyze the average `dpd` for different loan `status` categories.


**Reasoning**:
Perform the requested 4-point analysis on the merged dataset. This involves grouping by 'gender' and 'status' to calculate descriptive statistics, calculating the correlation between 'tenure' and 'amount', and calculating the mean 'dpd' by 'status'. Finally, print the results with appropriate headings.



```python

gender_analysis = merged_df.groupby('gender')[['amount', 'tenure', 'dpd']].agg(['mean', 'median', 'std', 'min', 'max'])

status_analysis = merged_df.groupby('status')[['amount', 'tenure', 'dpd']].agg(['mean', 'median', 'std', 'min', 'max'])

tenure_amount_correlation = merged_df['tenure'].corr(merged_df['amount'])

average_dpd_by_status = merged_df.groupby('status')['dpd'].mean()

print("--- Analysis by Gender ---")
display(gender_analysis)

print("\n--- Analysis by Status ---")
display(status_analysis)

print("\n--- Correlation between Tenure and Amount ---")
print(f"Correlation Coefficient: {tenure_amount_correlation:.4f}")

print("\n--- Average DPD by Status ---")
display(average_dpd_by_status)
```

```python
merged_df['dpd'] = pd.to_numeric(merged_df['dpd'], errors='coerce')
merged_df['is_late'] = merged_df['dpd'] > 0

if 'age' not in merged_df.columns:
    merged_df['dob'] = pd.to_datetime(merged_df['dob'], errors='coerce')
    merged_df['age'] = (pd.to_datetime('today') - merged_df['dob']).dt.days // 365


print(merged_df['is_late'].mean())

print(merged_df.groupby('gender')['is_late'].mean())
merged_df['age_group'] = pd.cut(merged_df['age'], bins=[18,25,35,45,60,80], labels=["18–25","26–35","36–45","46–60","60+"])
merged_df.groupby('age_group')['is_late'].mean()


```

## Data visualization

### Subtask:
Create visualizations to represent the findings from the data analysis. Specifically, visualize the distributions and relationships explored in the 4-point analysis.


**Reasoning**:
Import the necessary libraries for visualization.



```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

**Reasoning**:
Create visualizations to show the distribution of amount, tenure, and dpd across different gender categories using box plots.



```python
# 2. Visualize distribution of amount, tenure, and dpd across gender
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.boxplot(x='gender', y='amount', data=merged_df, ax=axes[0])
axes[0].set_title('Distribution of Loan Amount by Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Loan Amount')

sns.boxplot(x='gender', y='tenure', data=merged_df, ax=axes[1])
axes[1].set_title('Distribution of Loan Tenure by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Loan Tenure (Months)')

sns.boxplot(x='gender', y='dpd', data=merged_df, ax=axes[2])
axes[2].set_title('Distribution of DPD by Gender')
axes[2].set_xlabel('Gender')
axes[2].set_ylabel('Days Past Due (DPD)')

plt.tight_layout()
plt.show()
```

**Reasoning**:
Generate visualizations to display the distribution of amount, tenure, and dpd based on the status of the loan using box plots.



```python
# 3. Visualize distribution of amount, tenure, and dpd based on loan status
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.boxplot(x='status', y='amount', data=merged_df, ax=axes[0], order=['active', 'closed', 'default'])
axes[0].set_title('Distribution of Loan Amount by Status')
axes[0].set_xlabel('Loan Status')
axes[0].set_ylabel('Loan Amount')

sns.boxplot(x='status', y='tenure', data=merged_df, ax=axes[1], order=['active', 'closed', 'default'])
axes[1].set_title('Distribution of Loan Tenure by Status')
axes[1].set_xlabel('Loan Status')
axes[1].set_ylabel('Loan Tenure (Months)')

sns.boxplot(x='status', y='dpd', data=merged_df, ax=axes[2], order=['active', 'closed', 'default'])
axes[2].set_title('Distribution of DPD by Status')
axes[2].set_xlabel('Loan Status')
axes[2].set_ylabel('Days Past Due (DPD)')

plt.tight_layout()
plt.show()
```

**Reasoning**:
Create a scatter plot to visualize the relationship between tenure and amount, and a bar plot to show the average dpd for each loan status category.



```python
# 4. Scatter plot for relationship between tenure and amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tenure', y='amount', data=merged_df, alpha=0.6)
plt.title('Relationship between Loan Tenure and Amount')
plt.xlabel('Loan Tenure (Months)')
plt.ylabel('Loan Amount')
plt.show()
```
